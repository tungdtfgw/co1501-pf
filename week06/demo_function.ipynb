```python
# define a simple function with no arguments
def greet():
    print("Hello, World!")

# invoke / call the function
greet()
greet()
```

```python
# a simple function with an argument
def greet(name):
    print('Hello', name, '!')

# invoke / call the function, passing parameter values
greet('Alice')
greet('Bob')
```

```python
# a function with 2 arguments
def greet(first_name, last_name):
    name = f'{first_name} {last_name}'
    print('Hello', name, '!')

greet('Alice', 'Smith')
greet('Bob', 'Johnson')
```

```python
# a function that returns a value
def double(x):
    return x * 2


a = double(5)
print(a)

print(double(10))
```

```python
def add(x, y):
    z = x + y
    return z

a = add(5, 10)
print(a)
b = add(3, 6) * add(2, 10)
print(b)
```

```python
# write a function that returns rank by grade
def rank(grade):
    if grade < 40:
        result = 'Fail'
    elif grade <= 60:
        result = 'Pass'
    elif grade <= 80:
        result = 'Merit'
    else:
        result = 'Distinction'

    return result

grade = 75
print(rank(grade))

print(rank(35))
```
