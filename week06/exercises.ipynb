```python
def max3(a, b, c):
    m = a
    if b > m:
        m = b
    if c > m:
        m = c
    return m

print(max3(5, 10, 3))
print(max3(5, 3, 10))
print(max3(10, 5, 3))
```

```python
def is_square(x):
    if x < 0:
        return False
    
    root = int(x ** 0.5)
    return root * root == x

print(is_square(9))
print(is_square(8))
```
