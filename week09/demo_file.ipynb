```python
# Exercise 1: Read the content of cities.txt and count how many cities are in the file.
try:
    f = open('cities.txt', 'r')
    cities = f.read()
    cities = cities.split('\n')
    n_cities = len(cities)
    print(f'There are {n_cities} cities in the file.')
    for i in range(n_cities):
        print(f'{i+1}. {cities[i]}')
except FileNotFoundError:
    print('File not found!')
```

```python
try:
    f = open('cities.txt', 'r')
    cities = f.readlines()  # read all lines of the file
    n_cities = len(cities)
    print(f'There are {n_cities} cities in the file.')
    for i in range(n_cities):
        print(f'{i+1}. {cities[i].strip()}')    # use strip to remove \n character
except FileNotFoundError:
    print('File not found!')
```

```python
# read line by line
try:
    cities = []
    with open('cities.txt', 'r') as f:
        for line in f:  # read a line
            cities.append(line.strip()) # remove \n character
    n_cities = len(cities)
    print(f'There are {n_cities} cities in the file.')
    for i in range(n_cities):
        print(f'{i+1}. {cities[i]}')
except FileNotFoundError:
    print('File not found!')
```

```python
products = ['apple', 'banana', 'orange']
prices = [10, 20, 30]

try:
    f = open('products.txt', 'w')
    for i in range(len(products)):
        f.write(f'{products[i]} - {prices[i]}\n')
    f.close()   # close the file to push content from buffer to the file
except FileNotFoundError:
    print('File not found!')
print('All products saved.')
```
