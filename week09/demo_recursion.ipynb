```python
# recursive function is a function that calls itself
def hello():
    print('Hello, world!')
    answer = input('Do you want to say hello again? (y/n): ')
    if answer == 'y':
        hello()
```

```python
hello()
```

```python
def hello(n):
    if n <= 0: return
    print('Hello, world!')
    hello(n-1) # call hello again with n-1    
```

```python
hello(10)
```

```python
def sum(n):
    # base case (stop condition)
    if n <= 0: return 0
    # recursive case (call itself)
    return n + sum(n-1)

# Test
print(sum(5))
print(sum(10))
```

```python
def factorial(n):
    # base case (stop condition)
    if n <= 0: return 1 # 0! = 1
    # recursive case (call itself)
    return n * factorial(n-1)

# Test
print(factorial(5))
print(factorial(10))
```

```python
def fibonacci(n):
    if n == 0 or n == 1: return 1
    return fibonacci(n-1) + fibonacci(n-2)

# Test
for n in range(11):
    print(f'F({n}) = {fibonacci(n)}')
```
