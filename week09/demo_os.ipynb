```python
import os
```

```python
print(os.curdir) # current directory
print(os.pardir) # parent directory
```

```python
files = os.listdir(os.curdir) # get a list of files in the current directory
for f in files:
    print(f)
```

```python
files = os.listdir(os.pardir) # get a list of files in the parent directory
for f in files:
    print(f)
```

```python
files = os.listdir(os.curdir) # get a list of files in the current directory
for f in files:
    if os.path.isdir(f):
        print(f'+ {f}')
    else:
        print(f'- {f}')
```

```python
path = input('Enter a path: ')
# check if the path exist
if not os.path.exists(path):
    print(f'No such file or directory: {path}')
else:
    if os.path.isfile(path):    # if path is a file, print the file name
        print(f'- {path}')
    else:                       # if path is a directory, print its content
        for f in os.listdir(path):
            if os.path.isdir(path + '/' + f):
                print(f'+ {f}')
            else:
                print(f'- {f}')
```
