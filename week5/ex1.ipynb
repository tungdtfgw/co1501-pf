```python
# 1 số gọi là số nguyên tố nếu nó chỉ chia hết cho 1 và chính nó.
# Viết chương trình nhập vào n, lặp i từ 1 đến n, kiểm tra xem i có phải là số nguyên tố hay không, nếu có thì in ra i.
```

```python
n = int(input("Nhập vào n: "))

for i in range(2, n + 1):
    prime = True
    for j in range(2, i):
        if i % j == 0:
            prime = False

    if prime:
        print(i, end=" ")
```
