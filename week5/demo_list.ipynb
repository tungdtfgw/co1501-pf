```python
# Loop with condition
import random
numbers = random.sample(range(1, 100), 20)
print(numbers)
```

```python
for n in numbers:
    if n % 2 == 0:
        print(n, end=" ")
```

```python
count = 0
for n in numbers:
    if n % 2 == 0:
        count += 1

print(f'There are {count} even numbers in the list.')
```

```python
# Đếm có bao nhiêu số lẻ trong danh sách numbers
count = 0
for n in numbers:
    if n % 2 == 1:
        count += 1

print(f'There are {count} odd numbers in the list.')
# Đếm có bao nhiêu số lớn hơn 50 trong danh sách numbers
count = 0
for n in numbers:
    if n > 50:
        count += 1
print(f'There are {count} numbers greater than 50 in the list.')
```

```python
max_num = numbers[0]
for n in numbers:
    if n > max_num:
        max_num = n

print(f'The largest number in the list is {max_num}.')
```

```python
# Tìm min_num trong danh sách numbers
min_num = numbers[0]
for n in numbers:
    if n < min_num:
        min_num = n

print(f'The smallest number in the list is {min_num}.')
```

```python
x = int(input("Enter a number: "))
found = False
for n in numbers:
    if n == x:
        found = True

if found:
    print(f'{x} is in the list.')
else:
    print(f'{x} is not in the list.')
```

```python
pos = -1
x = int(input("Enter a number: "))
for i in range(len(numbers)):
    if numbers[i] == x:
        pos = i

if pos == -1:
    print(f'{x} is not in the list.')
else:
    print(f'{x} is in the list at position {pos}.')
```

```python
sum = 0
for n in numbers:
    sum += n

print(f'The sum of the numbers in the list is {sum}.')
```

```python
# Tính tổng các số chẵn trong danh sách numbers
sum_even = 0
for n in numbers:
    if n % 2 == 0:
        sum_even += n
print(f'The sum of the even numbers in the list is {sum_even}.')

# Tính tổng các số lẻ trong danh sách numbers
sum_odd = 0
for n in numbers:
    if n % 2 == 1:
        sum_odd += n
print(f'The sum of the odd numbers in the list is {sum_odd}.')
```

```python
# List slicing: get sub-list of a list
a = ["apple", "banana", "cherry", "kiwi", "mango", "orange", "grape", "watermelon"]
print(a, len(a))
```

```python
# slice the first 3 elements
b = a[0:3]
print(b)
```

```python
# slice from cherry to orange
c = a[2:6]
print(c)
```

```python
# slice the first 3 elements
b = a[:3]   # skip 0 because it is the default start index
print(b)
```

```python
# slice the last 3 elements
d = a[5:8]
print(d)
```

```python
# slice the last 3 elements
d = a[5:]  # skip 8 because it is the default end index
print(d)
```

```python
# slice middle elements (skip 2 first and 2 last elements)
e = a[2:6]
print(e)
```

```python
e = a[2:-2]  # skip 6 because it is the default end index
print(e)
```

```python
a = random.sample(range(1, 100), 20)
print(a)
```

```python
# slice elements with step
f = a[0:20:2]
print(f)
```

```python
f = a[::2]
print(f)
```

```python
g = a[::] # copy the whole list
print(g)
```

```python
b = a
c = a[::]

print(a, b, c, sep="\n")
```
