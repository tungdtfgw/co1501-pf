```python
# Exercise 1:
# Write a program that ask the user to enter height and weight
# Then calculate the BMI and print it to the user
# BMI = weight / height^2
# After that, print the BMI category based on the following:
# Underweight: BMI < 18.5
# Normal weight: 18.5 <= BMI < 25
# Overweight: 25 <= BMI < 30
# Obesity: BMI >= 30
```

```python
weight = float(input("Enter your weight in kg: "))
height = float(input("Enter your height in meters: "))
print(f"Your weight is: {weight} kg")
print(f"Your height is: {height} m")

BMI = weight / (height ** 2)
print(f"Your BMI is: {BMI:.2f}")

if BMI < 18.5:
    state = "Underweight"
elif BMI < 25:
    state = "Normal weight"
elif BMI < 30:
    state = "Overweight"
else:
    state = "Obesity"

print(f"You are {state}")
```

```python
# Exercise 2:
# Một công ty có chính sách thưởng cuối năm cho nhân viên như sau:
- Nếu là CEO: thưởng 10% lợi nhuận công ty hoặc 12 tháng lương, tùy theo giá trị nào lớn hơn
- Nếu là Manager: thưởng 6 tháng lương
- Nếu là Employee: thưởng 2 tháng lương
- Nếu là Employee có thâm niên > 5 năm: thưởng thêm 1 tháng lương

Viết chương trình yêu cầu người dùng nhập:
- Chức vụ (CEO, Manager, Employee)
- Lợi nhuận công ty
- Lương tháng
- Thâm niên (năm)

Sau đó tính và in ra số tiền thưởng cuối năm của nhân viên đó.
```

```python
position = input("Enter your position (CEO, Manager, Employee): ")
profit = float(input("Enter the company's profit: "))
salary = float(input("Enter your monthly salary: "))
experience = int(input("Enter your years of experience: "))

print(f"Your position is: {position}")
print(f"The company's profit is: {profit}")
print(f"Your monthly salary is: {salary}")
print(f"Your years of experience is: {experience}")

if position == "CEO":
    bonus = profit * 0.1
    if bonus < salary * 12:
        bonus = salary * 12
elif position == "Manager":
    bonus = salary * 6
else:
    bonus = salary * 3 if experience > 5 else salary * 2

print(f"Your year-end bonus is: {bonus:.2f}")
```

```python
Công thức tính tiền điện lũy kế:
- Dưới 50 kWh: 1678 đồng/kWh
- Từ 50 đến 100 kWh: 1734 đồng/kWh
- Từ 100 đến 200 kWh: 2014 đồng/kWh
- Trên 200 kWh: 2536 đồng/kWh

Nếu là khách hàng công ty, sẽ được giảm 10% trên tổng tiền điện.

Viết chương trình yêu cầu người dùng nhập:
- Số kWh tiêu thụ
- Loại khách hàng (Cá nhân hoặc Công ty)
Sau đó tính và in ra số tiền điện phải trả sau khi áp dụng chính sách giảm giá (nếu có).
```

```python
kwh = float(input("Enter the amount of kWh consumed: "))
cus_type = input("Enter customer type (Individual or Company): ")

if kwh < 50:
    cost = kwh * 1678
elif kwh < 100:
    cost = 50 * 1678 + (kwh - 50) * 1734
elif kwh < 200:
    cost = 50 * 1678 + 50 * 1734 + (kwh - 100) * 2014
else:
    cost = 50 * 1678 + 50 * 1734 + 100 * 2014 + (kwh - 200) * 2536

if cus_type == "Company":
    cost *= 0.9

print(f"The total electricity bill is: {cost:.2f} VND")
```
