```python
# conditional statement simple syntax
# if condition:
#     statement

full_price = 10

age = int(input("Enter your age: "))

if age < 18:
    full_price = full_price * 0.5

print(f"The price is: {full_price}")
```

```python
PRICE = 50

n_nights = int(input("Enter the number of nights: "))

if n_nights > 5:
    n_nights -= 1
    print("You get one free night!")

total_price = n_nights * PRICE
print(f"The total price is: {total_price}")
```

```python
grade = float(input("Enter your grade: "))
print(f"Your grade is: {grade}")

if grade < 40:
    print("You have failed the exam.")
elif grade <= 60:
    print("You passed the exam")
elif grade < 80:
    print("You passed the exam with good grade")
else:
    print("You passed the exam with excellent grade")
```

```python
age = int(input("Enter your age: "))

if age <= 6:
    print('Bus price for children is free')
elif age <= 22:
    print('Bus price for students is 5000 VND')
elif age < 60:
    print('Bus price for adults is 7000 VND')
else:
    print('Bus price for seniors is free')
```

```python
price = int(input("Enter the price: "))
distance = int(input("Enter the distance: "))
customer_type = input("Enter the customer type (vip / normal): ")

print(f"Price: {price}, Distance: {distance}, Customer type: {customer_type}")
```

```python
# Solution 1
if distance < 5:
    transport_fee = 0
elif customer_type == 'vip':
    transport_fee = 0
else:
    transport_fee = distance * 5000

total_price = price + transport_fee
print(f"The total price is: {total_price}")
```

```python
# Solution 2:
if distance < 5 or customer_type == 'vip':
    transport_fee = 0
else:
    transport_fee = distance * 5000

total_price = price + transport_fee
print(f"The total price is: {total_price}")
```
