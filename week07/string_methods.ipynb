```python
s = 'hello world'
print(s.islower())  # True
print(s.isupper())  # False
print(s.isalpha())  # False (contains space)
print(s.isdigit())  # False (non of the characters are digits)
```

```python
s = 'helloworld123'
print(s.isalnum())  # True (contains letters and digits)
print(s.isdigit())  # False (contains letters
print(s.isalpha())  # False (contains digits)
```

```python
s1 = 'hello world'
s2 = 'hello'
s3 = 'world'
print(s1.startswith(s2))  # True
print(s1.endswith(s3))  # True
print(s1.find(s2))  # 0 (index of the first occurrence of s2 in s1)
print(s1.find(s3))  # 6 (index of the first occurrence of s3 in s1)
```

```python
s4 = 'python'
print(s1.startswith(s4))  # False
print(s1.endswith(s4))  # False
print(s1.find(s4))  # -1 (s4 is not found in s1)
```

```python
s = 'hello, world'
print(s.count('o'))  # 2 (number of occurrences of 'o' in s)
print(s.count('l'))  # 3 (number of occurrences of 'l' in s)
print(s.count('z'))  # 0 (number of occurrences of 'z' in s)
```
