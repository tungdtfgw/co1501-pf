```python
# format integer numbers
a = 123
print('|123456789|')
print(f'|{a:9d}|')
```

```python
# format string
s = 'hello'
print('|123456789|')
print(f'|{s:9s}|')
```

```python
# format float
f = 3.14159
print('|123456789|')
print(f'|{f:9f}|')
print(f'|{f:9.2f}|')
print(f'|{f:9.2e}|')
print(f'|{f:9.2%}|')
```

```python
# alignment in format string
# > right alignment
# < left alignment
# ^ center alignment
a = 123
print('|123456789|')
print(f'|{a:>9d}|')
print(f'|{a:<9d}|')
print(f'|{a:^9d}|')
```

```python
student_ids = ["GCH1001", "GCH1002", "GCH1003"]
names = ['Adam Sandler', 'Brad Pitt', 'Cameron Diaz']
grades = [8.5, 9.0, 7.5]

def print_student(i):
    print(f'|{student_ids[i]:<10}|{names[i]:<20}|{grades[i]:>8.2f}|')

def print_all():
    print(f'|{'ID':^10}|{'Name':^20}|{'Grade':^8}|')
    print('-' * 42)
    for i in range(len(student_ids)):
        print_student(i)
        print('-' * 42)
```

```python
print_all()
```

```python
employees = ['John Smith', 'Paul Allen', 'Nancy Reed']
salaries = [45000, 78000, 65000]
departments = ['Accounting', 'IT', 'HR']
# TODO: print a table of employees with salaries and departments
# Extra: print the average salary of all employees

def print_employee(i):
    print(f'|{employees[i]:<20}|{salaries[i]:>10,.2f}|{departments[i]:<15}|')

def print_all_employees():
    print(f'|{'Employee Name':^20}|{'Salary':^10}|{'Department':^15}|')
    print('-' * 50)
    for i in range(len(employees)):
        print_employee(i)
        print('-' * 50)

def average_salary():
    total = 0
    for s in salaries:
        total += s
    
    avg = total / len(salaries)
    return avg
```

```python
print_all_employees()
print(f'Average Salary: {average_salary():.2f}')
```
