```python
a = 10
print(a)
```

```python
b = 2
print(b)
```

```python
# Interger's operators: +, -, *, /, //, %, **
print(a + b)  # Addition
print(a - b)  # Subtraction
print(a * b)  # Multiplication
print(a / b)  # Division (returns a float)
print(a // b)  # Floor division (returns an integer)
```

```python
print(10 % 3)  # Modulo (returns the remainder of the division)
print(10 % 2)
print(5 % 3)
```

```python
print(2 ** 3)  # Exponentiation (2 raised to the power of 3)
print(3 ** 2)
print(16 ** 0.5)  # Square root of 16
```

```python
# String's operators: +, *, in, not in
s1 = "Hello"
s2 = "World"
print(s1 + s2)  # Concatenation
print(s1 * 3)  # Repetition
print("lo" in s1)  # Substring check (True)
print("z" in s1)  # Substring check (False)
print("lo" not in s1)  # Substring check (False)
```
