```python
valid = False

while not valid:
    n = int(input('Enter a positive number: '))
    
    if n > 0:
        valid = True

print(f'You entered: {n}')
```

```python
limit = 500
total = 0

while total < limit:
    n = int(input('Enter a number: '))
    total += n
    print(f'Total so far: {total}')

print(f'Final total: {total}')
```

```python
answer = ''

while answer != 'quit':
    answer = input('Enter a command (or "quit" to exit): ')
    print(f'You entered: {answer}')
```

```python
import random as rd

n = rd.randint(1, 100)

incorrect = True

while incorrect:
    number = int(input('Guess a number between 1 and 10: '))
    print(f'You guessed: {number}')
    if number == n:
        print('Correct!')
        incorrect = False
    elif number < n:
        print('Too low, try again.')
    else:
        print('Too high, try again.')
```
