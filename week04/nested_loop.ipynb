```python
print_table = True

while print_table:
    # Get n from user
    n = int(input('Enter n: '))
    # print multiplication table for n
    for i in range(10):
        print(f'{n} x {i + 1} = {n * (i + 1)}')
    # Ask user if they want to print another table
    answer = input('Print another table? (y/n): ')
    if answer != 'y':
        print_table = False
```

```python
n = int(input('Enter n: '))

for row in range(1, n + 1):
    for i in range(row):
        print('* ', end='')
    print()
```

```python
print_triangle = True

while print_triangle:
    n = int(input('Enter n: '))

    for row in range(1, n + 1):
        for i in range(row):
            print('* ', end='')
        print()

    answer = input('Print another triangle? (y/n): ')
    print_triangle = answer == 'y'
```
