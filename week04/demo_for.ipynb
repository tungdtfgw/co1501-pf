```python
# Simple for syntax for n times: 
# for i in range(n):
#    do something
```

```python
# print the string 'Hello' 5 times
for i in range(5):
    print('Hello')
```

```python
for i in range(5):
    print('Hello', end=' ')
```

```python
# print numbers
for i in range(5):
    print(i, end=' ')
```

```python
for i in range(5):
    print(i+1, end=' ')
```

```python
for i in range(1, 6):
    print(i, end=' ')
```

```python
# Ex1: print numbers: 1 3 5 7 ... 19
# C1:
for i in range(10):
    print(2*i+1, end=' ')

print()
# C2:
for i in range(20):
    if i % 2 == 1:
        print(i, end=' ')

print()
# C3:
for i in range(1, 20, 2):
    print(i, end=' ')
```

```python
# Ex2: print numbers: 2 4 6 8 ... 20
for i in range(10):
    print(2*(i+1), end=' ')

print()

for i in range(1, 21):
    if i % 2 == 0:
        print(i, end=' ')

print()

for i in range(2, 21, 2):
    print(i, end=' ')
```
